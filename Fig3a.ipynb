```julia
## Line plot 
    # Range of x values
        x_values = -4:0.005:4
        y =1.0
        xtick_positions = -4.0:1.0:4.0
        xtick_labels = ["-4","-3","-2","-1","0","1","2","3","4"]
        ytick_positions = [0.0,0.05]
        ytick_labels = ["0","0.05"]
        xlabel_size = 16
        ylabel_size = 16
        ticklabel_size = 12
    # Create the plot
        p = plot()
        y_values_for_plot = [U_of_x_y_quadgk(x, y) for x in x_values]
        max_y = maximum(y_values_for_plot)
        y_values_for_plot = 0.05 * y_values_for_plot/max_y
        plot!(p, x_values, y_values_for_plot, linecolor=:black, linewidth=2)
        xlabel!("\$(x_I-x_J)/  \$")
        ylabel!("\$U_{IJ}/v\$")
        plot!(p, 
            legend=false,
            size = [350,300],
            ylim=[-0.003,0.06],
            xlim=[-4,4],
            xticks=(xtick_positions, xtick_labels),
            yticks=(ytick_positions, ytick_labels),
            tick_direction=:in,
            grid = false,
            xlabelfontsize=xlabel_size,
            ylabelfontsize=ylabel_size,
            xtickfontsize=ticklabel_size, 
            ytickfontsize=ticklabel_size
        )

## Scatter plot
    ind = 1::Int
    avals = [3.5,2.5,1.0,0.8]
    vals = [0.4,0.4,0.4,0.4]
    color_theme = palette(:vik, 2000)
    colorsss = vcat(RGB{Float64}(1.0,1.0,1.0),color_theme[800],color_theme[1200],color_theme[600])
    nu = vals[ind]
    a = avals[ind]
    lB = 1.0
     = 1.0
    N=101
    x = range(0.0, 1, length=10^5)
    y = cos.(2*x)  # make sure it's periodic w/ period 1
    wire_positions, _ = find_wire_positions_and_velocities(x,y,nu)
    dd_couplings, dd_positions = density_density_couplings(N,wire_positions,a,lB,,true)
    dd_positions2=dd_positions
    dd_couplings = 0.05 * dd_couplings/max_y
    [dd_positions2[j] = -dd_positions[j] for j in 1:N]
    scatter!(dd_positions2,dd_couplings,
        marker = (:circle, 8, 1.0, colorsss[ind], stroke(2, 1.0, :black, :dot)),
        legend = false)
```
